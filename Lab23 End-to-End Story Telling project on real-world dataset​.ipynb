##  Data Cleaning
Check for missing values and handle them appropriately.

Correct any inconsistencies in the data.

Ensure all data types are appropriate.

```python
# Check for missing values
print(df.isnull().sum())

# Fill missing values or drop them if necessary
df.fillna(method='ffill', inplace=True)

# Check data types
print(df.dtypes)
```

## Data Visualization
Use visualizations to understand the distribution and relationships between variables.
Create histograms, box plots, scatter plots, bar charts, and heatmaps.

```python
# Visualizations
plt.figure(figsize=(12, 10))

# Gold medals distribution
plt.subplot(221)
df['gold'].plot(kind='hist', bins=10, color='gold', alpha=0.7, title='Gold Medals Distribution')
plt.xlabel('Number of Gold Medals')

# Silver medals distribution
plt.subplot(222)
df['silver'].plot(kind='hist', bins=10, color='silver', alpha=0.7, title='Silver Medals Distribution')
plt.xlabel('Number of Silver Medals')

# Bronze medals distribution
plt.subplot(223)
df['bronze'].plot(kind='hist', bins=10, color='#cd7f32', alpha=0.7, title='Bronze Medals Distribution')
plt.xlabel('Number of Bronze Medals')

# Total medals distribution
plt.subplot(224)
df['total'].plot(kind='hist', bins=10, color='blue', alpha=0.7, title='Total Medals Distribution')
plt.xlabel('Total Medals')

plt.tight_layout()
plt.show()
```

Observations: Histograms show the distribution of gold, silver, bronze, and total medals among countries.

```python
num_cols = ['gold', 'silver', 'bronze', 'total', 'gdp', 'population']
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
numeric_df = df[num_cols].copy()

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

The heatmap reveals strong correlations between total medals and individual medal types. GDP and population also correlate with medal counts.

```python
# Scatter plot: GDP vs Total Medals
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdp', y='total', size='population', hue='region', data=df, alpha=0.7)
plt.title('GDP vs Total Medals')
plt.xlabel('GDP (in billion USD)')
plt.ylabel('Total Medals')
plt.legend(title='Region')
plt.show()
```

A scatter plot indicates countries with higher GDPs tend to win more medals. The bubble size reflects population.

```python
# Grouped analysis: Average medals by region
region_grouped = df.groupby('region')[['gold', 'silver', 'bronze', 'total']].mean().sort_values('total', ascending=False)
print(region_grouped)
region_grouped.plot(kind='bar', figsize=(12, 8), title='Average Medals by Region', colormap='viridis')
plt.ylabel('Average Number of Medals')
plt.xticks(rotation=45)
plt.show()
```

The bar chart shows the average number of medals by region, with Europe and Asia leading in medal counts.

## Descriptive Statistics
Calculate and interpret measures of central tendency and dispersion.
Summarize key statistics for numerical variables.

```python
# Descriptive statistics
print(df.describe())

mean_total_medals = df['total'].mean()
median_total_medals = df['total'].median()
std_total_medals = df['total'].std()

print(f"Mean Total Medals: {mean_total_medals}")
print(f"Median Total Medals: {median_total_medals}")
print(f"Standard Deviation of Total Medals: {std_total_medals}")

```

## Outlier Detection
Box Plot for Outlier Detection in Total Medals<br>
The box plot shows the distribution of total medals and identifies potential outliers.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['total'])
plt.title('Box Plot for Total Medals')
plt.show()

# Handling Outliers (Using IQR Method)
q1 = df['total'].quantile(0.25)
q3 = df['total'].quantile(0.75)
iqr = q3 - q1

# Define outliers
outliers = df[(df['total'] < (q1 - 1.5 * iqr)) | (df['total'] > (q3 + 1.5 * iqr))]
print("Outliers in Total Medals:")
print(outliers)
```
